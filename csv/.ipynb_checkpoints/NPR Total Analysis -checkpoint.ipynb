```python
import pandas as pd
df = pd.read_csv('NPR.Total.csv')
df = df.set_index('Show')
df = df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
import re
df['Body'] = df['Body'].apply(lambda x: re.sub('[^\w\s]', '', x))
```
